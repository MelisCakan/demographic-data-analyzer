### What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K? What percentage of people without advanced education make more than 50K?

```python
higher_education = df[(df.education == 'Bachelors') | (df.education == 'Masters') | (df.education == 'Doctorate')]
lower_education = df[(df.education != 'Bachelors') & (df.education != 'Masters') & (df.education != 'Doctorate')]
```

```python
higher_education_rich = round((higher_education[higher_education["salary"] == ">50K"].shape[0] / higher_education.shape[0]) * 100, 1)
```

```python
lower_education_rich = round((lower_education[lower_education["salary"] == ">50K"].shape[0] / lower_education.shape[0]) * 100, 1)
```

```python
higher_education_rich
```

```python
lower_education_rich
```

### What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = df[(df["hours-per-week"] == df["hours-per-week"].min()) & (df["salary"] == ">50K")].shape[0]
num_min_workers
```

```python
round(num_min_workers / df.shape[0] * 100, 1)
```

### What country has the highest percentage of people that earn >50K?

```python
df.head()
```

```python
total_people_by_country = df["native-country"].value_counts()
rich_people_by_country = df[df["salary"] == ">50K"]["native-country"].value_counts()
percentage_by_country = (rich_people_by_country / total_people_by_country * 100)
highest_earning_country = percentage_by_country.idxmax()
highest_percentage = round(percentage_by_country.max().item(), 1)
```

```python
highest_earning_country
```

```python
highest_percentage
```
